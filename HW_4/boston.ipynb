```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
from sklearn import datasets
housedata = datasets.load_boston()
housedata
```

```python
housedata.keys()
```

```python
house_features = housedata.feature_names
```

```python
data = pd.DataFrame(housedata.data, columns = house_features)
data.head()
```

```python
house_target = pd.DataFrame(housedata.target, columns = ['PRICE'])
house_target.head()
```

```python
data['PRICE'] = house_target
data.head()
```

```python
data.isnull().sum()
```

```python
df_corr  = data.corr()
```

```python
plt.figure(figsize = (10,10))
sns.set(font_scale = 0.8)
sns.heatmap(df_corr, annot = True, cbar = True)
plt.show()
```

```python
# 2x4 subplot . axs 4x2 
fig, axs = plt.subplots(figsize=(16,4), ncols=4, nrows=1)

lm_features = ['RM','INDUS','PTRATIO','LSTAT']

# i  feature RM ~ RAD  
for i, feature in enumerate(lm_features):
    # 2
    col = i
    # sns.regplot :  
    sns.regplot(x=feature, y='PRICE', data=data, ax=axs[col])
```

```python
X_data = data[lm_features]
X_data.head()
```

```python
y_data = data['PRICE']
y_data.head()
```

```python
# 0. EDA, 1. , 2.Scaler, 3. , 4.      
```

```python
train_input, test_input, train_target, test_target = train_test_split(
    X_data, y_data, test_size=0.2, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
print ("==================================")
print (" ACCURACY OF OTHER ML REGRESSION" )
print ("==================================")
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(train_scaled, train_target)
print(lr.coef_ ,lr.intercept_ ) # , y
```

```python
print(lr.score(train_scaled, train_target))
print(lr.score(test_scaled, test_target))
```

```python
y_pred = lr.predict(test_scaled)
y_pred
```

```python
test_input.shape
```

```python
c = [i for i in range(1,len(test_target)+1,1)]
sns.scatterplot(test_input['RM'] , y = y_pred, color = ['red'] , label = ['pred'])
sns.scatterplot(test_input['RM'], y = test_target, color = ['blue'] , label = ['target'])
#plt.plot(c, lr.predict(test_scaled), color='blue')
#plt.plot(test_input['RM'], lr.coef_[0] * test_input['RM'] + lr.coef_[1] * test_input['INDUS'] + lr.coef_[2] * test_input['PTRATIO'] 
#         + lr.coef_[3]* test_input['LSTAT'] + lr.intercept_)
plt.title('prediction_target')
plt.show()
```

```python
c = [i for i in range(1,len(test_target)+1,1)]
plt.plot(c,test_target - y_pred, color='green',linestyle='-')
plt.xlabel('index')
plt.ylabel('Error')
plt.title('Error Value')
plt.show()
```
